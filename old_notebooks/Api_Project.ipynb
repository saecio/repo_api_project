```python
import requests
import pandas as pd
from dotenv import load_dotenv
import os
import time 

load_dotenv()  

app_id = os.getenv("app_id")   
app_key = os.getenv("app_key")   

url = "https://api.adzuna.com/v1/api/jobs/es/search/1"

params = {
    "app_id": app_id,
    "app_key": app_key,
    "what": "data analyst",
    #"where": "Milan",
    "results_per_page": 20,
    "content-type": "application/json"
}

response = requests.get(url, params=params)
data = response.json()
```

```python
# 2B- Check the status
if response.status_code == 200:
    print("Connection successful!")
else:
    print(f"Connection failed: {response.status_code}")
```

```python
# All European countries Adzuna covers
european_countries = [
    "at",    # Austria
    "be",    # Belgium
    "fr",    # France
    "de",    # Germany
    "it",    # Italy
    "nl",    # Netherlands
    "pl",    # Poland
    "es",    # Spain
    "ch",    # Switzerland
    "gb",    # United Kingdom
]
```

```python
all_jobs = []
```

```python
#To get more jobs than 1 page as is only limited by 50 per page
for country in european_countries:
    print(f"\nFetching jobs for {country}...")
    page = 1

    while True:
        url = f"https://api.adzuna.com/v1/api/jobs/{country}/search/{page}"

        params = {
            "app_id": app_id,
            "app_key": app_key,
            "what": "data analyst",
            "results_per_page": 50,
            "content-type": "application/json"
        }

        response = requests.get(url, params=params)

        if response.status_code != 200:
            print(f"✗ {country} page {page}: {response.status_code}")
            break

        data = response.json()
        results = data.get("results", [])

        if not results:
            print(f"✓ {country}: no more jobs found after page {page - 1}")
            break

        for job in results:
            job["country"] = country
            job["page"] = page

        all_jobs.extend(results)
        print(f"✓ {country} page {page}: {len(results)} jobs")

        page += 1
        time.sleep(0.5)


raw_jobs_df = pd.DataFrame(all_jobs)
print(f"\nTotal: {len(raw_jobs_df)} jobs from {len(european_countries)} countries")
raw_jobs_df.head()
```

```python
from bs4 import BeautifulSoup

pd.set_option("display.max_colwidth", None)
pd.set_option("display.max_columns", None)

def get_full_job_description(url, delay=0.2):
    if pd.isna(url):
        return pd.NA

    try:
        response = requests.get(
            url,
            timeout=10,
            headers={"User-Agent": "Mozilla/5.0"},
        )
        response.raise_for_status()
    except requests.RequestException:
        return pd.NA
    finally:
        time.sleep(delay)

    soup = BeautifulSoup(response.text, "html.parser")

    for tag in soup(["script", "style", "noscript"]):
        tag.decompose()

    full_description = " ".join(soup.get_text(" ").split())
    return full_description if full_description else pd.NA


raw_jobs_df["full_description"] = raw_jobs_df["redirect_url"].apply(get_full_job_description)

raw_jobs_df["description"] = raw_jobs_df["full_description"].combine_first(
    raw_jobs_df["description"]
)

raw_jobs_df = raw_jobs_df.drop(columns=["full_description"])

raw_jobs_df
```
