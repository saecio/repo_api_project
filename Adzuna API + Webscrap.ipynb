# Using adzuna API + Scrapping in it

```python
# ============================================================
# NOTEBOOK RUNNER
# Executes the three pipeline scripts in order with live output.
# ============================================================

import subprocess
import sys

scripts = [
    "2_adzuna_extract.py",
    "3_adzuna_scrape.py",
    "4_adzuna_export.py"
]

for script in scripts:
    print(f"\n{'='*50}", flush=True)
    print(f"Running {script}...", flush=True)
    print('='*50, flush=True)

    result = subprocess.run(
        [sys.executable, "-u", script],
        text=True
    )

    if result.returncode != 0:
        print(f"✗ ERROR in {script}", flush=True)
        break

    print(f"✓ {script} completed successfully", flush=True)

print("\n✓ All done!", flush=True)
```
